```python
r = requests.get('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
soup = BeautifulSoup(r.text,'html.parser')
```

```python
T=[]
tickers = soup.find_all('a',{'class':'external text'},{'rel':'nofollow'})
for i in tickers[:-180]:
    if i.text == 'reports':
        del i
    else:
        T.append(i.text)
```
